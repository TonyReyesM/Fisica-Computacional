**Ejercicio 1**

Calcular el área de un círculo y un elipse, y el volumen de una esfera y un cilindro circular.

*Ecuación del área de un círculo: $A = π r^2$*


```
# Calculamos el área de un CÍRCULO
# Definimos la función del área de círculo con el radio como argumento
def areaCirc(rad):
  return (np.pi)*rad**2
```

```
# Le pedimos al usuario el valor del radio
print('Ingrese el valor del radio ', end='')
r = float(input())
# Calculamos el valor del área al introducir el valor del radio
aCirc = areaCirc(r)
```

```
# Imprime el valor del área en formato libre.
#print("\nArea = ", aCirc)
# Imprime el valor del área con 2 decimales. 
print("\nArea = {:.2f}".format(aCirc))
```

*Ecuación del área de un elipse: $A = abπ$*

```
# Calculamos el área de un ELIPSE
# Definimos la función del área de elipse con los semiejes como argumento
def areaElips(a, b):
  return (np.pi)*a*b
```

```
# Le pedimos al usuario el valor de cada semieje
print('Ingrese el valor del primer semieje ', end='')
x = float(input())
print('Ingrese el valor del segundo semieje ', end='')
y = float(input())
# Calculamos el valor del área al introducir el valor de los semiejes
aElips = areaElips(x, y)
```

```
# Imprime el valor del área en formato libre.
#print("\nArea = ", aElips)
# Imprime el valor del área con 2 decimales. 
print("\nArea = {:.2f}".format(aElips))
```

*Ecuación del volumen de una esfera: $V = (4/3)πr^3$*

```
# Calculamos el volumen de una ESFERA
# Definimos la función del volumen de la esfera con el radio como argumento
def volEsf(rad):
  return (4/3)*(np.pi)*rad**3
```

```
# Le pedimos al usuario el valor del radio
print('Ingrese el valor del radio ', end='')
rEsf = float(input())
# Calculamos el valor del volumen al introducir el valor del radio
vEsf = volEsf(rEsf)
```

```
# Imprime el valor del volumen en formato libre.
#print("\nVolumen = ", vEsf)
# Imprime el valor del volumen con 2 decimales. 
print("\nVolumen = {:.2f}".format(vEsf))
```

*Ecuación del volumen de un cilindro circular: $V = hπr^2$*



```
# Calculamos el volumen de un CILINDRO CIRCULAR
# Definimos la función del volumen del cilindro con el radio y la altura como argumento
def volCil(rad, y):
  return y*(np.pi)*rad**2
```

```
# Le pedimos al usuario el valor del radio y la altura
print('Ingrese el valor del radio ', end='')
rCil = float(input())
print('Ingrese el valor de la altura ', end='')
h = float(input())
# Calculamos el valor del volumen al introducir el valor del radio y la altura
vCil = volCil(rCil, h)
```

```
# Imprime el valor del volumen en formato libre.
#print("\nVolumen = ", vCil)
# Imprime el valor del volumen con 2 decimales. 
print("\nVolumen = {:.2f}".format(vCil))
```

Desarrollar un programa que calcule las raíces de una ecuación cuadrática.

*Ecuación cuadrática: $ax^2 + bx + c$*

Fórmula general: $\frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $

```
# Importamos la biblioteca necesaria para usar números complejos
import cmath
```

```
# Crearemos una función para calcular las raíces de una ecuación cuadrática
# Definimos la función con los coeficientes de la ecuación como argumentos
def solRaiz(a, b, c):
  raiz = b**2 - 4*a*c
  if raiz >= 0:
    raiz1 = (-b + np.sqrt(raiz))/(2*a)
    raiz2 = (-b - np.sqrt(raiz))/(2*a)
  else:
    raiz = -raiz
    raizx1 = -b/(2*a)
    raizy1 = +np.sqrt(raiz)/(2*a)
    raizx2 = -b/(2*a)
    raizy2 = -np.sqrt(raiz)/(2*a)
    raiz1 = complex(raizx1, raizy1)
    raiz2 = complex(raizx2, raizy2)
  return raiz1, raiz2
  
```

```
# Obtenemos los valores de los coeficientes a b y c del usuario
print('Ingrese el valor del coeficiente cuadrático:')
cuadrat = float(input())
print('Ingrese el valor del 120coeficiente lineal:')
lin = float(input())
print('Ingrese el valor de la constante:')
const = float(input())
```

```
# Imprimimos las raíces
raices = solRaiz(cuadrat, lin, const)

print('Las raíces son: ', raices)
```

**Ejercicio 3**

Calcular la raíz de un número arbitrario utilizando el método de Heron

*Método de Heron:*

*   $x_0\approx \sqrt{S}$
*   $x_{n+1} = \frac{1}{2}(x_n + \frac{S}{x_n})$
*   $\sqrt{S} = \ \lim_{n\to\infty} x_n \$





```
# Calculamos la raíz cuadrada de un número arbitrario usando el metodo de Heron
# Definimos la función en la que habrá una iteración del método de Heron que correrá hasta obtener un resultado con un error menor a 0.01
def metHeron(x_0):
  i = 1
  err = 1
  while (err > 0.01):
    heron = (i +(x_0/i))/2
    err = abs((np.sqrt(x_0)-heron)/np.sqrt(x_0))
    i = heron
  return heron
```

```
# Pedimos al usuario el número del cual se quiere calcular la raíz
print('Ingrese el número del cual desea encontrar la raíz:')
c = float(input())
```

```
# Evaluamos el valor obtenido del usuario en la funcion del método de Heron
#metHeron(aprox)
# Imprimimos el valor para el usuario
print('La raíz de ', c,' por el método de Heron es: ', metHeron(c))
print('La raíz exacta de', c,'es: ', np.sqrt(c))
```

Reproducir la figura que aparece inmediatamente abajo en el artículo de Wikipedia sobre Series de Taylor, que muestra la aproximación de la función $ln(1+x)$ alrededor de $x=0$.

*Serie de Taylor para $ln(1 + x)$:*


*   $ln(1 + x) = \
\sum_{n=1}^{\infty}(-1)^{n+1} \frac{x^n}{n}=x_{1}+x_{2}+x_{3} 
\$



```
# Crearemos una gráfica para la función ln(x+1) en la que se hacen aproximaciones utilizando la serie de Taylor alrededor
# de x=0
# Se graficará la serie de Taylor para la función con 4, 7 , 11 y 16 términos junto a la función de ln(x+1)
```

```
# Creamos un arreglo de 100 puntos entre -1 y 1 en el eje x y un arreglo para los valores de la función evaluada en cada 
# punto
x = np.linspace(-2, 2, 100)
y = np.log(x+1)
```

```
# Creamos arreglos para los primeros 16 términos de la serie de Taylor de ln(x+1) alrededor de x=0
y1 = x
y2 = -(x**2)/2
y3 = (x**3)/3
y4 = -(x**4)/4
y5 = (x**5)/5
y6 = -(x**6)/6
y7 = (x**7)/7
y8 = -(x**8)/8
y9 = (x**9)/9
y10 = -(x**10)/10
y11 = (x**11)/11
y12 = -(x**12)/12
y13 = (x**13)/13
y14 = -(x**14)/14
y15 = (x**15)/15
y16 = -(x**16)/16
```

```
# Creamos los arreglos para las series de Taylos de 4,7, 11, y 16 términos
sT4 = y1 + y2 + y3 + y4
sT7 = y1 + y2 + y3 + y4 + y5 + y6 + y7
sT11 = y1 + y2 + y3 + y4 + y5 + y6 + y7 + y8 + y9 + y10 + y11
sT16 = y1 + y2 + y3 + y4 + y5 + y6 + y7 + y8 + y9 + y10 + y11 + y12 + y13 + y14+ y15 + y16
```

```
# Graficamos las cinco funciones en el intervalo -1 < x < 1 en una misma figura
plt.plot(x, y, label = 'ln(x+1)')
plt.plot(x, sT4, label = 'T4')
plt.plot(x, sT7, label = 'T7')
plt.plot(x, sT11, label = 'T11')
plt.plot(x, sT16, label = 'T16')
plt.grid()

# Colocamos un título a la figura
plt.title('Series de Taylor: ln(x+1)')

# Definimos los límites del eje x y eje y
plt.xlim(-1.5,1.5)
plt.ylim(-4,2)

# Ponemos las leyendas de cada una de las curvas
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
```
